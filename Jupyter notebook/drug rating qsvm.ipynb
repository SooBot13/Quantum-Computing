```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('drugdata.csv')
df=pd.DataFrame(df)
df=df.head(100)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
# df.State=le.fit_transform(df['State'])
df.drugName=le.fit_transform(df['drugName'])
df.condition=le.fit_transform(df['condition'])
```

```python
df
```

```python
df['consideration'] = [1 if x > 5 else 0 for x in df['rating']]        
```

```python
df
```

```python
X=df[['drugName', 'condition', 'rating', 'usefulCount'] ].to_numpy()
X = np.asarray(X, dtype = 'int')
y=df['consideration'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.svm import SVC
svc_model=SVC()
svc_model.fit(X_train, y_train)
```

```python
train_score_c4 = svc_model.score(X_train, y_train)
test_score_c4 = svc_model.score(X_test, y_test)

print(f"Classical SVC on the training dataset: {train_score_c4:.2f}")
print(f"Classical SVC on the test dataset:     {test_score_c4:.2f}")
```

```python
y_pred_svc=svc_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm_svc=confusion_matrix( y_test, y_pred_svc )
sns.heatmap(cm_svc,
            annot=True,
            fmt='g',
            xticklabels=['Yes','No'],
            yticklabels=['Yes','No'])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
from qiskit.circuit.library import ZZFeatureMap

num_features = X.shape[1]

feature_map = ZZFeatureMap(feature_dimension=num_features, reps=1)
feature_map.decompose().draw(output="mpl", fold=20)
```

```python
from qiskit.circuit.library import RealAmplitudes

ansatz = RealAmplitudes(num_qubits=num_features, reps=3)
ansatz.decompose().draw(output="mpl", fold=20)
```

```python
from qiskit.algorithms.optimizers import COBYLA

optimizer = COBYLA(maxiter=100)

from qiskit.primitives import Sampler

sampler = Sampler()
```

```python
from matplotlib import pyplot as plt
from IPython.display import clear_output

objective_func_vals = []
plt.rcParams["figure.figsize"] = (12, 6)


def callback_graph(weights, obj_func_eval):
    clear_output(wait=True)
    objective_func_vals.append(obj_func_eval)
    plt.title("Objective function value against iteration")
    plt.xlabel("Iteration")
    plt.ylabel("Objective function value")
    plt.plot(range(len(objective_func_vals)), objective_func_vals)
    plt.show()
```

```python
import time
from qiskit_machine_learning.algorithms.classifiers import VQC

vqc1 = VQC(
    sampler=sampler,
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=optimizer,
    callback=callback_graph,
)

# clear objective value history
objective_func_vals = []

start = time.time()
vqc1.fit(X_train, y_train)
elapsed = time.time() - start

print(f"Training time: {round(elapsed)} seconds")
```

```python
train_score_q4 = vqc1.score(X_train, y_train)
test_score_q4 = vqc1.score(X_test, y_test)

print(f"Quantum VQC on the training dataset: {train_score_q4:.2f}")
print(f"Quantum VQC on the test dataset:     {test_score_q4:.2f}")
```

```python
from qiskit.circuit.library import ZZFeatureMap
from qiskit.primitives import Sampler
from qiskit.algorithms.state_fidelities import ComputeUncompute
from qiskit_machine_learning.kernels import FidelityQuantumKernel
```

```python
feature_map = ZZFeatureMap(feature_dimension=2, reps=2, entanglement="linear")

sampler = Sampler()

fidelity = ComputeUncompute(sampler=sampler)

drug_kernel = FidelityQuantumKernel(fidelity=fidelity, feature_map=feature_map)
```

```python
from qiskit_machine_learning.algorithms import QSVC
import time

qsvc = QSVC(quantum_kernel=drug_kernel)

start = time.time()

qsvc.fit(X_train, y_train)

qsvc_score = qsvc.score(X_test, y_test)

elapsed = time.time() - start

print(f"QSVC classification test score: {qsvc_score}")

print(f"Training time: {round(elapsed)} seconds")
```
